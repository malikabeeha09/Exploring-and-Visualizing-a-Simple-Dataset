```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\hp\Downloads\Iris.csv')

print("Shape of dataset:", df.shape)
print("Column names:\n", df.columns)
print("First 5 rows of the dataset:\n", df.head())
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Id', y='Species', data=df)
plt.title("Scatter Plot of ColumnX vs ColumnY")
plt.xlabel("ColumnX")
plt.ylabel("ColumnY")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Species'], bins=20, kde=True)
plt.title("Histogram of species")
plt.xlabel("species")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Species'])
plt.title("Box Plot of ColumnZ")
plt.ylabel("ColumnZ")
plt.show()
```
